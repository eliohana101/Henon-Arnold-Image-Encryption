```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
# Create data

img = cv2.imread('ISIC_0024318.png') 
x = img[0:254,0:254,0]
# print(x)
y = img[1:255,1:255,0]
area = np.pi*10

# Plot
plt.scatter(x, y, s=area, alpha=0.5)
plt.title('Scatter plot pythonspocom')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
