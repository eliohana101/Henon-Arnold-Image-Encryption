```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
# Create data

img = cv2.imread('ISIC_0024306.jpg')
print(img.shape)
x = img[0:img.shape[0]-1,0:img.shape[1]-1,0]
# print(x)
y = img[1:img.shape[0],0:img.shape[1]-1,0]
area = 0.01

# Plot
plt.scatter(x, y, s=area, alpha=0.5)
plt.title('Horizontal pixel correlation')
plt.xlabel('pixel value at (x,y)')
plt.ylabel('pixel value at (x+1,y)')
plt.show()
```

```python
x = img[0:img.shape[0]-1,0:img.shape[1]-1,0]
# print(x)
y = img[0:img.shape[0]-1,1:img.shape[1],0]
area = 0.01

# Plot
plt.scatter(x, y, s=area, alpha=0.5)
plt.title('Vertical pixel correlation')
plt.xlabel('pixel value at (x,y)')
plt.ylabel('pixel value at (x,y+1)')
plt.show()
```

```python
x = img[0:img.shape[0]-1,0:img.shape[1]-1,0]
# print(x)
y = img[1:img.shape[0],1:img.shape[1],0]
area = 0.01

# Plot
plt.scatter(x, y, s=area, alpha=0.5)
plt.title('Diagonal pixel correlation')
plt.xlabel('pixel value at (x,y)')
plt.ylabel('pixel value at (x+1,y+1)')
plt.show()

print(x[0,0],y[0,0])
```
